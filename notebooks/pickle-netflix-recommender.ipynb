```python
import pickle 
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df = df[(df['type'] == 'Movie') & (df['country'] == 'United States')]
df = df.reset_index(drop=True)
df.shape
```

```python
#import vectorizer and model
tfidf = pickle.load(open("vect_01.pkl", "rb"))
nn = pickle.load(open("knn_01.pkl", "rb"))
```

```python
movie = ["vampire"]
```

```python
new = tfidf.transform(movie)
nn.kneighbors(new.todense())
recommendations = nn.kneighbors(new.todense())[1].tolist()[0]
df[['type','title','cast','listed_in','description']].iloc[recommendations]
```
